```
# Import pandas. Pandas is a very useful library to work with and to manipulate data. Much of my time in data science is actually working with this particular library when examining and cleaning data sets 

import pandas as pd
from google.colab import files
#uploaded = files.upload()
!wget https://raw.githubusercontent.com/bennng/data/master/automobile.csv
```

```
# I set some missing values to look out for in my data set and then I read the file with read_csv. 
# Once read I print out the number of records that has the missing values I set with the variable missing_values.
# Note the other code I commented out which deal with alternative ways to grab the data file.
missing_values = ["NaN","n/a", "na", "--", "?"]
#import io
#df = pd.read_csv(io.BytesIO(uploaded['automobile.csv']),  na_values = missing_values)
df = pd.read_csv('automobile.csv',  na_values=missing_values)
#print (df['Num-of-doors'])
#print (df['Num-of-doors'].isnull())
print (df.isnull().sum())
```

```
#Cleaning data: In most data sets you find you will always run into the problem of missing data.
#Always research the data. I see what I can do to fill in the missing data. Sometimes  I talk with stake holders. Sometime i might talk to other SMEs. 
# Sometime I will try to find the missing data from another datasource, perhaps a different database. 
#Case in point here we are looking at data dealing with automobiles. If there is missing data about the number of doors I might just go to the car makers website and look up the car. 
# I would then proceed to fill out that data if I can. If there are no other choices you can delete or use the mean, median or mode of that set.

doorsmode = df['Num-of-doors'].mode()
doorsmode
df['Num-of-doors'].fillna(doorsmode, inplace=True)

boremedian = df['Bore'].median()
boremedian
df['Bore'].fillna(boremedian, inplace=True)

strokemedian = df['Stroke'].median()
strokemedian
df['Stroke'].fillna(strokemedian, inplace=True)

hpmedian = df['Horsepower'].median()
hpmedian
df['Horsepower'].fillna(hpmedian, inplace=True)

rpmode = df['Peak-rpm'].median()
rpmode
df['Peak-rpm'].fillna(rpmode, inplace=True)

pricemedian = df['Price'].median()
pricemedian
df['Price'].fillna(pricemedian, inplace=True)

df.loc[4,'Num-of-doors'] = 'four'
df.loc[171,'Num-of-doors'] = 'four'

print (df.isnull().sum())
#print(df['Make'].unique())

```

```
# I don't like '-' in my data set so I modify the header to not have them

makedummies = pd.get_dummies(df.Make)
makemergeddf = pd.concat([df, makedummies], axis='columns')
dropdummytrapmakemergeddf = makemergeddf.drop(['Make','alfa-romero'], axis='columns')
#dropdummytrapmakemergeddf

dropdummytrapmakemergeddf.rename(columns = {'Fuel-type':'Fueltype'}, inplace = True)
dropdummytrapmakemergeddf.rename(columns = {'Num-of-doors':'Doors'}, inplace = True)
dropdummytrapmakemergeddf.rename(columns = {'Body-style':'Body'}, inplace = True)
dropdummytrapmakemergeddf.rename(columns = {'Drive-wheels':'Wheeldrive'}, inplace = True)
dropdummytrapmakemergeddf.rename(columns = {'Engine-location':'Enginelocation'}, inplace = True)
dropdummytrapmakemergeddf.rename(columns = {'Wheel-base':'Wheelbase'}, inplace = True)
dropdummytrapmakemergeddf.rename(columns = {'Curb-weight':'Curbweight'}, inplace = True)
dropdummytrapmakemergeddf.rename(columns = {'Engine-type':'Enginetype'}, inplace = True)
dropdummytrapmakemergeddf.rename(columns = {'Num-of-cylinders':'Numofcylinders'}, inplace = True)
dropdummytrapmakemergeddf.rename(columns = {'Engine-size':'Enginesize'}, inplace = True)
dropdummytrapmakemergeddf.rename(columns = {'Fuel-system':'Fuelsystem'}, inplace = True)
dropdummytrapmakemergeddf.rename(columns = {'Compression-ratio':'Compressionratio'}, inplace = True)
dropdummytrapmakemergeddf.rename(columns = {'Peak-rpm':'Peakrpm'}, inplace = True)
dropdummytrapmakemergeddf.rename(columns = {'City-mpg':'Citympg'}, inplace = True)
dropdummytrapmakemergeddf.rename(columns = {'Highway-mpg':'Highwaympg'}, inplace = True)

dropdummytrapmakemergeddf.rename(columns = {'mercedes-benz':'mercedesbenz'}, inplace = True)

dropdummytrapmakemergeddf
```

```
from sklearn.tree import DecisionTreeRegressor
tree_reg = DecisionTreeRegressor()
tree_reg.fit(X,Y)
```

```
import numpy as np

from sklearn.metrics import mean_squared_error
symboling_predictions = tree_reg.predict(X)
tree_mse = mean_squared_error(Y, symboling_predictions)
tree_rmse = np.sqrt(tree_mse)
tree_rmse
```
